```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
url = 'student_data.csv'
df = pd.read_csv(url)

# Display the major subject-wise student count
subject_counts = df['Major'].value_counts()
print("Major subject-wise student count:\n", subject_counts)

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(subject_counts, labels=subject_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Major subject-wise student distribution')
plt.show()
```
